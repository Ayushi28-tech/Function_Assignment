1.What is the difference between a function and a method in Python?

### Function:

* A function is a block of code that performs a specific task and can be defined using the def keyword.
* Functions are independent of objects and can be called directly.
* Functions are global and can be used in any part of the program after they are defined.
* Functions are not tied to any particular object or class.
* Not associated with any specific object.
* Parameters are explicitly defined in the function signature.
* Called directly using its name.

   Example: function_name(arguments)

#### Example
        def greet(name):
            return f"Hello, {name}!"

        print(greet("Ayushi"))  # Output: Hello, Ayushi!

### Method:

* A method is a function that is associated with an object. It is called on the object and typically operates on the object's attributes.
* Methods are defined within a class and take the instance (usually self) as their first parameter.
* Methods are local to the object or class they are defined in and require an instance of the class to be called (unless it is a @staticmethod or @classmethod).
* Always associated with an object, and often operates on the object's attributes.
* The first parameter of a method is typically self, which refers to the instance of the class.
* Called on an object using the dot (.) notation.

   Example: object_name.method_name(arguments)

#### Example
      class Greeter:
           def greet(self, name):
                return f"Hello, {name}!"

      greeter = Greeter()
      print(greeter.greet("Ayushi"))  # Output: Hello, Ayushi!

5.What are iterators in Python and how do they differ from iterables?

### Iterable:

* Definition: An object that can return its elements one at a time, allowing iteration over its contents.
* Examples: Lists, tuples, sets, dictionaries, strings.
* How it works: Uses the __iter__() method to return an iterator.

* Example:

     my_list = [1, 2, 3]
         for item in my_list:  # my_list is an iterable
    print(item)

### Iterator:

* Definition: An object that represents a stream of data, producing one element at a time on demand.
* How it works:

       Uses the __next__() method to retrieve the next item.
       Implements both __iter__() (returns itself) and __next__() methods.

* Example:

       my_list = [1, 2, 3]
       iterator = iter(my_list)  # Create an iterator from an iterable
       print(next(iterator))  # Output: 1
       print(next(iterator))  # Output: 2
       print(next(iterator))  # Output: 3








6.Explain the concept of generators in Python and how they are defined.

-> Generators are a type of iterable, like lists or tuples, but they produce values lazily, one at a time, using the yield keyword. Unlike lists, generators do not store all values in memory, making them memory-efficient for large datasets or infinite sequences.

* A generator function is defined like a normal function but uses the yield keyword instead of return.
* Example:

         def count_up_to(max_value):
              current = 1
              while current <= max_value:
                    yield current
                    current += 1

         #Using the generator

         for number in count_up_to(5):
               print(number)



7.What are the advantages of using generators over regular functions?

--> Generators offer several advantages, especially when dealing with large datasets, infinite sequences, or scenarios where memory and performance optimization are critical. Here's why they are preferred in certain cases:

### Memory Efficiency:

##### Generators:

* Produce one value at a time and do not store the entire dataset in memory.
* Ideal for large datasets or infinite sequences.

##### Regular Functions:

* Store all results in memory, which can lead to high memory usage for large datasets.

#### Example

         def generate_numbers():
             for i in range(1000000):  # Large range
                  yield i

         gen = generate_numbers()  # Only one value is stored at a time

### Infinite Sequences:

##### Generators:

* Can generate infinite sequences without exhausting memory.

##### Regular Functions:

* Cannot handle infinite sequences as they require all values to be stored in memory.

#### Example
      
         def infinite_counter():
             n = 1
             while True:
                  yield n
                  n += 1

### Readability and Maintainability:

##### Generators:

* Make code cleaner and more concise for iterating over complex data sequences.

##### Regular Functions:

* May require additional logic for iteration and state management.

### Improved Performance:

##### Generators:

* Yield values one at a time, making them faster for use cases where only a subset of data is required.

##### Regular Functions:

* Return all values at once, which can slow down performance for large computations.

9.Explain the purpose and usage of the `map()` function in Python.

--> The map() function in Python is a built-in higher-order function that applies a given function to all items in an iterable (such as a list, tuple, etc.) and returns a map object (which is an iterator) of the results. It allows for functional programming-style operations where a function is applied to each element in a collection.

#### Syntax
        map(function, iterable, ...)

* function: The function to apply to each item in the iterable(s).
* iterable: The iterable(s) whose elements the function will process. Multiple iterables can be passed, and the function will apply to corresponding elements in the iterables.
* The map() function returns an iterator, so it needs to be converted to a list, tuple, or another collection to view the results.

#### Example

        def add(x, y):
            return x + y

        list1 = [1, 2, 3]
        list2 = [4, 5, 6]
        result = map(add, list1, list2)
        print(list(result))  # Output: [5, 7, 9]





10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

### map():

##### Purpose:

The map() function applies a given function to each item in an iterable and returns an iterator with the results.

##### Syntax:
       map(function, iterable)

##### Example:
         numbers = [1, 2, 3, 4]
         squared = map(lambda x: x**2, numbers)
         print(list(squared))  # Output: [1, 4, 9, 16]

### filter():

##### Purpose:

The filter() function filters elements of an iterable based on a predicate (a function that returns True or False) and returns an iterator with only the elements that satisfy the condition.

##### Syntax:
        filter(function, iterable)

##### Example:
        numbers = [1, 2, 3, 4, 5]
        even_numbers = filter(lambda x: x % 2 == 0, numbers)
        print(list(even_numbers))  # Output: [2, 4]

### reduce():

##### Purpose:

The reduce() function applies a rolling computation to the items of an iterable, using a binary function (a function that takes two arguments). Unlike map() and filter(), reduce() reduces the iterable to a single accumulated value.

##### Syntax:
        from functools import reduce
        reduce(function, iterable)

##### Example:
        
        from functools import reduce
        numbers = [1, 2, 3, 4]
        sum_result = reduce(lambda x, y: x + y, numbers)
        print(sum_result)  # Output: 10

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            even_sum += num
    return even_sum

numbers_list = [10, 15, 20, 25, 30]
result = sum_of_even_numbers(numbers_list)
print("Sum of all even numbers in the list:", result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
    if num < 2:  # Numbers less than 2 are not prime
        return False
    for i in range(2, int(num ** 0.5) + 1):  # Check divisors up to the square root of the number
        if num % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
      print(number, "is a prime number.")
```

```python
#5. . Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    """
    Iterator class to generate the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1  # Initialize the first two Fibonacci numbers

    def __iter__(self):
        """
        Returns the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Generate the next number in the Fibonacci sequence.
        """
        if self.current_term >= self.num_terms:
            raise StopIteration  # End the iteration when the limit is reached
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            fib = self.a + self.b
            self.a, self.b = self.b, fib  # Update for the next term
            self.current_term += 1
            return fib

# Example usage:
fib_iterator = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for fib_num in fib_iterator:
    print(fib_num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent  # Yield the power of 2

for power in powers_of_two(5):
    print(power, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Strip newline characters from each line
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
file_name = "example.txt"  # Replace with the actual file path
for line in read_file_line_by_line(file_name):
    print(line)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples
tuples_list = [(1, 5), (3, 2), (4, 8), (2, 4)]

# Sort the list of tuples by the second element using a lambda function
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Print the sorted list
print("Original List:", tuples_list)
print("Sorted List:", sorted_list)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    """
    Convert a temperature from Celsius to Fahrenheit.

    :param celsius: Temperature in Celsius
    :return: Temperature in Fahrenheit
    """
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]

# Use map() to convert each Celsius temperature to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the results
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string

def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, World! Python is amazing."

# Use filter() to remove vowels
filtered_string = "".join(filter(is_not_vowel, input_string))

# Print the results
print("Original String:", input_string)
print("String without Vowels:", filtered_string)
```
